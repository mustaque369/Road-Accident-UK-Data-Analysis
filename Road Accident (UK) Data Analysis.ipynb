# Road Accident (UK) Dataset Analysis
The UK government amassed traffic data from 2000 and 2015, recording over 1.8 million accidents in the process and making this one of the most comprehensive traffic data sets out there. It's a huge picture of a country undergoing change.

###  Description
* This Dataset contains 33 Features covering over a 1.8 million records. this dataset consists of various features like, Accidental / Longitude: Location of Accident

* Accident_Severity: Accident Severity on the Scale of 1 to 5

* Number_of_Vehicles: Number of Vehicles Involved

* Number_of_Casualties: Number of Casualties in Accident

* Light_Conditions: Lighting Condition on the day of Accident

* Weather_Conditions: Weather Conditions on the day of the Accident

* Road_Surface_Conditions: Road Surface Conditions of Accidental Spot

* Year: Year of Accidental Event

### Surface Road Conditon

```python
df.Road_Surface_Conditions
```

```python
s_condition=df.Road_Surface_Conditions.unique()
len(s_condition)
```

```python
accidents_by_weather=df.Road_Surface_Conditions.value_counts()
accidents_by_weather
```

```python
c=['red','yellow','orange','blue','green']
accidents_by_weather[:5].plot(kind='barh',color=c,alpha=0.5)
plt.xlabel('Accidents',fontsize=13)
plt.ylabel('Road surface condition',fontsize=13)
plt.title('Road condition',fontsize=19)
plt.show()
```

The maximum number of accidents occurs in dry road conditions highlights the complexity of driver behavior and the role of perceived safety. It underscores the need for a holistic approach to road safety that includes education, infrastructure improvements, and leveraging technology. Addressing this issue can contribute significantly to reducing accidents and making roadways safer for all.

### Day of week

```python
df.Day_of_Week
```

```python
most_accident_week_day=df.Day_of_Week.value_counts()
most_accident_week_day
```

```python
most_accident_week_day.plot(kind='bar',color='orange',alpha=0.7)
plt.xlabel('Week',fontsize=15)
plt.ylabel('Accidents',fontsize=15)
plt.title('Accidents Per Week',fontsize=18)
plt.show()
```

### Time

```python
df.Time=pd.to_datetime(df.Time)
sns.histplot(df.Time.dt.hour,bins=24,color='r',alpha=0.5)
plt.ylabel('Accidents count',fontsize=15)
plt.xlabel('Time',fontsize=15)
plt.title('Time of Accident',fontsize=18)
plt.show()
```

The graph shows that a significant trend where a considerable number of accidents happen during two specific time slots: 7am to 9am and 3pm to 5pm. These time periods correspond to busy traffic times when people are often going to work or returning home. 


### Year

```python
accident_per_year=df.Year.value_counts()
accident_per_year
```

```python
accident_per_year.plot(kind='line',color='b',alpha=0.5)
plt.ylabel('Accidents',size=15)
plt.xlabel('Year',size=15)
plt.title('Accidents Per Year',size=18)
plt.show()
```

__The analysis of accident data spanning from 2005 to 2014 uncovers a consistent and positive trend:__ the number of accidents has decreased over these years. This finding suggests that efforts aimed at improving road safety, enhancing driver awareness, and implementing stricter traffic regulations may have yielded tangible results.

### Speed Limit

```python
max_accident_speed_limit=df.Speed_limit.value_counts()
```

```python
c=['blue','orange','green','red','purple','brown']
max_accident_speed_limit.plot(kind='bar',alpha=0.6,color=c)
plt.ylabel('Accidents',size=15)
plt.xlabel('Speed Limit',size=15)
plt.title('Maximum Accident / Speed Limit',size=18)
plt.show()
```

* The maximum number of accidents occurred at the 30 km/h speed limit. This finding underscores the critical relationship between speed and road safety, highlighting the significance of speed limits in preventing accidents and mitigating their severity.

* The emphasis on the 30 km/h speed limit as a frequent context for accidents signals the need for a comprehensive reevaluation of road safety strategies. 

### Number of Vehicles Involved in the accident

```python
vehicles_involved=df.Number_of_Vehicles.value_counts()
vehicles_involved=vehicles_involved.head(4)
vehicles_involved
```

```python
ex=[0.05,0.0,0.0,0.0]
vehicles_involved.plot(kind='pie',explode=ex,autopct="%0.1f%%",startangle=180)
plt.title('Most Vehicles collided',size=18)
plt.legend(loc=1)
plt.ylabel(" ")
plt.show()
```

In the pie chart, it's clear that the most common reason for accidents is when two vehicles collide. This means that a lot of accidents happen when two cars or vehicles crash into each other.

### Light Conditions

```python
df['Light_Conditions']
```

```python
Light=df.Light_Conditions.value_counts()
Light
```

```python
#Modifying the terminology for the sake of simplicity.
df.loc[df['Light_Conditions'] == 'Daylight: Street light present', 'Light_Conditions'] = 'Daylight'
df.loc[df['Light_Conditions'] == 'Darkness: Street lights present and lit', 'Light_Conditions'] = 'Street Light On'
df.loc[df['Light_Conditions'] == 'Darkeness: No street lighting', 'Light_Conditions'] = 'No Street Light'
df.loc[df['Light_Conditions'] == 'Darkness: Street lighting unknown', 'Light_Conditions'] = 'Street lighting unknown'
df.loc[df['Light_Conditions'] == 'Darkness: Street lights present but unlit', 'Light_Conditions'] = 'Street lighting Off'
df['Light_Conditions']
```

```python
sns.scatterplot(x="Light_Conditions",y="Number_of_Casualties",data=df)
sns.set(rc={'figure.figsize':(17,6)})
plt.title("Light Condition",size=18)
plt.xlabel("Light",size=15)
plt.ylabel("Number of Casualties",size=15)
plt.show()
```

The most accidents resulting in injuries happened during the day when there's plenty of light. This tells us that road safety isn't just about nighttime or bad weather.  Even in good visibility conditions, accidents can still occur due to various reasons like more traffic on the road, driver distractions, or people feeling too comfortable because it's daytime. This discovery reminds us that road safety efforts are needed all the time, not just when it's dark or difficult to see. It's a reminder to stay cautious and alert throughout the day to prevent accidents and make roads safer for everyone.

## Insights
* The absence of city-specific information limits our ability to dive into localized accident trends.
* Accidents are more frequent on dry roads, followed by wet/damp surfaces. This highlights the importance of road maintenance and careful driving practices regardless of weather.
* Saturdays observe the highest accident rates, indicating the necessity of heightened safety awareness during leisure periods when traffic dynamics might differ.
* Accidents peak during 7am - 9am and 3pm - 5pm, likely due to higher traffic volume. Managing traffic flow and promoting safe driving behaviors during these times could be crucial.
* The consistent decrease in accidents from 2005 to 2014 suggests successful road safety measures during that period. Understanding these interventions could guide future strategies.
* The 30 km/h speed limit witnesses the most accidents, underscoring the significance of enforcing speed rules, especially in crowded or pedestrian-heavy zones.
* Collisions involving two vehicles are the most frequent. This emphasizes the need for educating drivers, better signage, and stricter traffic enforcement to prevent multi-vehicle accidents.
* Most accidents occur during daylight hours, raising questions about visibility and potential distractions or fatigue that need further exploration.

__These findings provide a roadmap for targeted interventions and informed decisions to enhance road safety.__
